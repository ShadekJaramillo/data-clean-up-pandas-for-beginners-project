#### Exercise 06. Remove the null values (NAs) from the dataset, if applicable (★★☆)

After removing the null values, compare the size of the DataFrame before and after the removal.

```python
# TODO

ds_no_null_in_columns = ds.dropna(axis=1)
ds_no_null_columns = ds.dropna(axis=1, how='all')
ds_no_null_in_rows = ds.dropna()
ds_no_null_rows = ds.dropna(how='all')

datasets = [ds, ds_no_null_in_columns, ds_no_null_columns, ds_no_null_in_rows, ds_no_null_rows]
shapes = [dataset.shape for dataset in datasets]
sizes = [dataset.size for dataset in datasets]
messages = [
    'shape of the original dataset',
    'shape of the dataset after dropping the columns that contain any null value',
    'shape of the dataset after dropping the null columns',
    'shape of the dataset after dropping the rows containing any null value',
    'shape of the dataset after dropping the null rows'
]

for dataset, shape, size, message in zip(datasets, shapes, sizes, messages):
    message_ending =  f'({ds.size-size} less than the original dataset)' if dataset is not ds else ''
    print(f'{message}: {shape}, size={size} {message_ending}')
```

#### Exercise 07. What is the average price in the population of "Arroyomolinos (Madrid)"? (★★☆)

Print the value obtained from the 'level5' column.

```python
# TODO

houses_in_Arroyomolinos_Madrid = ds.loc[ds.level5 == 'Arroyomolinos (Madrid)']
print(f'the average price of listed house in Arroyomolinos, Madrid (sale or rent or maybe even lease) is {houses_in_Arroyomolinos_Madrid.price.mean():,.2f} EUR')
```

#### Exercise 08. Plot the histogram of prices for the population of "Arroyomolinos (Madrid)" and explain what you observe (★★☆)

Print the histogram of the prices and write a brief analysis of the plot in the Markdown cell.

```python
import matplotlib.pyplot as plt
import numpy as np

max_val = houses_in_Arroyomolinos_Madrid.price.max()
min_val = houses_in_Arroyomolinos_Madrid.price.min()

# Plot the histogram of prices

plt.figure()
plt.title('House listings in Arroyomolinos (Madrid) price histogram')
plt.xlabel("price (EUR)")
plt.ylabel('number of house')
plt.hist(houses_in_Arroyomolinos_Madrid.price, bins = np.arange(min_val,max_val+1, step=50000, dtype=np.int64))
plt.show()

print(f'the average house listing price in Arroyomolinos (Madrid) is: {houses_in_Arroyomolinos_Madrid.price.mean():,.2f} EUR')
```

# **Plot analysis**

This plot resembles a normal distribution, which means that houses with prices around $\approx$ 300,000.00 EUR were more likely to be listed in the region of Arroyomolinos, Madrid on this webpage at the time of the data recollection.

The data is not enough to be completely sure that it is indeed a gaussian distribution, and there is somithing that points in the direction of alternative distributions, it is the gap between houses with prices of 150,000 and 50,000 EUR. This is an argument that may be stronger if take into account that some of the listings are for rents and maybe even listings, which means that there is a chance that the isle of houses with prices ranging between 0 - 50,000 EUR may actually be neglegible and is not relevant for the analysis of the house sale prices.
